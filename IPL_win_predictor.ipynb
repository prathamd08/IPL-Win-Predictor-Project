```python
ball_by_ball_df = matches_df.merge(ball_by_ball,on = 'ID')
```

```python
ball_by_ball_df =ball_by_ball_df[ball_by_ball_df['innings']==2]
```

```python
ball_by_ball_df
```

```python
ball_by_ball_df['current_score']= ball_by_ball_df.groupby('ID').cumsum()['total_run_y']
```

```python
ball_by_ball_df
```

```python
ball_by_ball_df['runs_left'] = ball_by_ball_df['total_run_x'] - ball_by_ball_df['current_score'] + 1
```

```python
ball_by_ball_df
```

```python
ball_by_ball_df['balls_left'] = 120 - (ball_by_ball_df['overs']*6 + ball_by_ball_df['ballnumber'])
```

```python
ball_by_ball_df
```

```python
ball_by_ball_df['wickets_left']= 10 - ball_by_ball_df.groupby('ID').cumsum()['isWicketDelivery']
```

```python
ball_by_ball_df.iloc[[27]]
```

```python
ball_by_ball_df['crr'] = (ball_by_ball_df['current_score']*6)/(120 - ball_by_ball_df['balls_left'] )
```

```python

ball_by_ball_df
```

```python
ball_by_ball_df['rrr'] = (ball_by_ball_df['runs_left']*6)/ball_by_ball_df['balls_left']
```

```python
def results(row) :
  return 1 if row['WinningTeam'] == row['BattingTeam'] else 0 ;
```

```python
ball_by_ball_df['results']=ball_by_ball_df.apply(results,axis=1)
```

```python
ball_by_ball_df
```

```python
final_df = ball_by_ball_df[['BattingTeam','BallingTeam','City','runs_left','balls_left','wickets_left','crr','rrr','results']]
```

```python
df1 = final_df.dropna()
```

```python
df1 = df1[df1['balls_left'] != 0]
```

```python
x = df1.iloc[:,:-1]
y = df1.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
is_infinite = np.isinf(df1['crr'])
has_infinite = is_infinite.any()
has_infinite
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BallingTeam','City'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
